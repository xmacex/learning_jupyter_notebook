What is a meshgrid, in Numpy, and also in Bokeh?

```python
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting
%matplotlib inline
```

```python
bokeh.plotting.output_notebook()
```

If we just take [this](http://bokeh.pydata.org/en/latest/docs/gallery/image.html) as a staring point

```python
N = 500
x = np.linspace(0, 10, N)
y = np.linspace(0, 10, N)
xx, yy = np.meshgrid(x, y)
d = np.sin(xx)*np.sin(yy)
```

```python
xx.shape, yy.shape
```

```python
p = bokeh.plotting.figure(x_range=(0, 10), y_range=(0, 10))
p.image(image=[d], x=0, y=0, dw=10, dh=10, palette="Spectral11")
bokeh.plotting.show(p)
```

And try to understand such a nice pattern via [*What is the purpose of meshgrid in Python*](https://stackoverflow.com/questions/36013063/what-is-purpose-of-meshgrid-in-python). Meshgrid is being used to plot decision boundaries for ML tasks. What meshgrid does, is takes two lists of numbers, and generates all the points, with element from each. Not unlike `zip`, in two dimensions.

```python
tx, ty = np.meshgrid(range(0, 5), range(0, 3))
tx, ty
```

```python
plt.plot(tx, ty, marker='.', linestyle='none')
```

Same with matplotlib's `matplotlib.pyplot.contourf`.

```python
plt.contourf(x, y, d)
```

```python
z = np.sin(xx**3 + yy**3) / (xx**2 + yy**2)
plt.contourf(x, y, z)
```

```python
def sinus2d(x, y):
    return np.sin(x) + np.sin(y)
```

Ok to plot the results of above field function in the range $0-2\pi$

```python
sx, sy = np.meshgrid(np.linspace(0, 2*np.pi, 100), np.linspace(0, 2*np.pi, 100))
sz = sinus2d(sx, sy)
```

With Bokeh

```python
sp = bokeh.plotting.figure(x_range=(0, 2*np.pi), y_range=(0, 2*np.pi))
sp.image(image=[sz], x=0, y=0, dw=2*np.pi, dh=2*np.pi, palette="Spectral3")
bokeh.plotting.show(sp)
```

And contours with Matplotlib

```python
plt.contourf(sx, sy, sz)
```
