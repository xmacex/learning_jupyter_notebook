# Plotting with Bokeh

Ok Matplotlib is a burden, let's face it. Some solutions include things like *seaborn*, *Altair*, *Holoviews*, and *Bokeh* which I'm learning here

```python
import bokeh
%matplotlib inline
```

Now, starting with the [Quickstart](http://bokeh.pydata.org/en/latest/docs/user_guide/quickstart.html#userguide-quickstart), a line plot. In Bokeh, high-level things are in interface `bokeh.plotting`, and low-level in interface `bokeh.models`, latter centered around visuals.

```python
import bokeh.plotting
```

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = bokeh.plotting.figure(title="Simple linechart", x_axis_label="x", y_axis_label="y")
p.line(x, y, legend="Nice data")
bokeh.plotting.show(p)
```

So the routine for the higher level plotting API is

1. prepare data
1. tell Bohek where to output, a HTML file or Notebook
1. call `figure()`
1. add renderers, such as `Figure.line`, `Figure.circle` and define visual customizations
1. show or save the result

For interaction in a Notebook there is [a demo](https://github.com/bokeh/bokeh/blob/0.12.6/examples/howto/notebook_comms/Jupyter%20Interactors.ipynb), which uses the `ipywidgets`. It defines an update function which does `bokeh.io.push_notebook()` after updating a `line.data_source.data['y']`, and then calls `interact` from the widgets, and passes the update function.

```python
import ipywidgets
import numpy as np
```

```python
interx = np.linspace(0, 2*np.pi, 2000)
intery = np.sin(interx)

bokeh.io.output_notebook()
interp = bokeh.plotting.figure(title="Interaction demo", plot_height=300, plot_width=600, y_range=(-5,5))
interr = interp.line(interx, intery, color="#2222aa", line_width=3)

def update(f, w=1, A=1, phi=0):
    if f == "sin": func = np.sin
    elif f == "cos": func = np.cos
    elif f == "tan": func = np.tan
    interr.data_source.data['y'] = A * func(w * interx + phi)
    bokeh.io.push_notebook()

bokeh.plotting.show(interp, notebook_handle=True)
ipywidgets.interact(update,
                    f=["sin", "cos", "tan"],
                   w=(0,100),
                   A=(1,5),
                   phi=(0, 20, 0.1))
```
