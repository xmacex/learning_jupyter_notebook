# Venn diagrams

Let's try some Venn diagrams, from the [`matplotlib_venn`](https://pypi.python.org/pypi/matplotlib-venn) package.

## Two sets with `venn2`

Function `venn2` draws, blotches colours and labels.

```python
animals = (12, 7, 3)
animals = [set(['cheetah', 'cat', 'lynx', 'shark', 'snake']), set(['tamagotchi', 'cat', 'undulate', 'snake'])]
matplotlib_venn.venn2(subsets=animals, set_labels=['wild', 'domesticated']) # for region sizes
```

Whereas `venn2_circles` draws only the circles.

```python
matplotlib_venn.venn2_circles(subsets=animals)
```

Both can be drawn on top of one another, and configured in a variety of ways. Also subplots via `matplotlib.pyplot.subplots` (not shown here).

```python
matplotlib.pyplot.figure(figsize=(6, 6))
ss2 = (3, 2, 1)
venn2patches, venn2circles = matplotlib_venn.venn2(subsets=ss2, set_labels=('some stuff', 'more stuff', 'shared stuff')), matplotlib_venn.venn2_circles(subsets=ss2, linestyle='dotted')
venn2patches.get_patch_by_id('100').set_color('blue')
venn2patches.get_label_by_id('100').set_text('colourful circles')
venn2patches.get_patch_by_id('110').set_color('green')
venn2patches.get_label_by_id('110').set_text('Venn\ndiagrams')
venn2patches.get_patch_by_id('010').set_color('yellow')
venn2patches.get_label_by_id('010').set_text('diagrams')
venn2circles[1].set_ls('dashed')
matplotlib.pyplot.title('A configured Venn diagram', size=14)
matplotlib.pyplot.annotate('This is what we\ntry to achieve',
                           xy=venn2patches.get_label_by_id('110').get_position() + np.array([0, 0.05]),
                           xytext=(-120, 50),
                           ha='center',
                           textcoords='offset points',
                           bbox=dict(boxstyle='round, pad=0.5',
                                     fc='green',
                                     alpha=0.3),
                           arrowprops=dict(arrowstyle='->',
                                           connectionstyle='arc3, rad=-0.4',
                                           color='red'))
matplotlib.pyplot.show()
```
